```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('trafikkdata.csv', sep=';')

data.head()
```

```python
# replace - with 0
data = data.replace('-', 0)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)

data[["Trafikkmengde"]]=imputer.fit_transform(data[["Trafikkmengde"]])
```

```python
# print row 448-460
print(data[447:470])
```

```python
data['Elgeseter NO2µg/m³'].plot()
```

```python
data['Elgeseter PM2.5µg/m³'].plot()
```

```python

data = data[['Fra-tid','Til-tid' , 'Elgeseter PM10µg/m³', 'Elgeseter PM2.5µg/m³', 'Elgeseter NO2µg/m³']]

data.head(20)

data.to_csv('luftkvalitet.csv', index=False)
```
