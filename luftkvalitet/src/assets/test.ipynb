```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('luftkvalitet.csv', sep=',', header=0)
data.head(20)


data['Elgeseter PM10µg/m³'].plot()
data['Elgeseter PM2.5µg/m³'].plot()
data['Elgeseter NO2µg/m³'].plot()
plt.show()
```

```python

# set data that is minus to 0

data.loc[data['Elgeseter PM10µg/m³'] < 0, 'Elgeseter PM10µg/m³'] = 0
data.loc[data['Elgeseter PM2.5µg/m³'] < 0, 'Elgeseter PM2.5µg/m³'] = 0
data.loc[data['Elgeseter NO2µg/m³'] < 0, 'Elgeseter NO2µg/m³'] = 0
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=10)

data[["Elgeseter PM10µg/m³", "Elgeseter PM2.5µg/m³", "Elgeseter NO2µg/m³"]] = imputer.fit_transform(data[["Elgeseter PM10µg/m³", "Elgeseter PM2.5µg/m³", "Elgeseter NO2µg/m³"]])
```

```python
data['Elgeseter PM2.5µg/m³'].plot()
```

```python
data['Elgeseter NO2µg/m³'].plot()
```

```python
data['Elgeseter PM2.5µg/m³'].plot()
```

```python

data = data[['Fra-tid','Til-tid' , 'Elgeseter PM10µg/m³', 'Elgeseter PM2.5µg/m³', 'Elgeseter NO2µg/m³']]

data.head(20)

data.to_csv('luftkvalitet.csv', index=False)
```
